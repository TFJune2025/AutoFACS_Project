```python
import os

def compare_folders(folder1, folder2, output_file):
    """
    Compare the contents of two folders and list the files that are unique to each folder.

    Args:
        folder1 (str): Path to the first folder.
        folder2 (str): Path to the second folder.
        output_file (str): Path to the output file where differences will be logged.
    """
    # Get the list of files in each folder
    files_in_folder1 = set(os.listdir(folder1))
    files_in_folder2 = set(os.listdir(folder2))

    # Find unique files in each folder
    unique_to_folder1 = files_in_folder1 - files_in_folder2
    unique_to_folder2 = files_in_folder2 - files_in_folder1

    # Write the results to the output file
    with open(output_file, "w") as f:
        f.write(f"Files unique to {folder1}:\n")
        for file in unique_to_folder1:
            f.write(f"{file}\n")
        
        f.write(f"\nFiles unique to {folder2}:\n")
        for file in unique_to_folder2:
            f.write(f"{file}\n")

    print(f"Comparison complete. Results saved to {output_file}")

# Example usage
folder1 = "lfw_dataset_curated"
folder2 = "lfw_dataset_curated_2"
output_file = "lfw_nofaces.txt"

compare_folders(folder1, folder2, output_file)
```
