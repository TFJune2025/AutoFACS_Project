```python
from PIL import Image
import os
```

```python
def count_total_images(dataset_path):
    total_images = 0

    for root, _, files in os.walk(dataset_path):
        for file in files:
            # Check for valid image file extensions
            if file.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tiff")):
                total_images += 1
    
    print("\nTotal number of images:", total_images)
    return total_images
```

```python
def check_image_sizes(dataset_path):
    image_sizes = {}  # Dictionary to store image size counts
    for root, _, files in os.walk(dataset_path):
        for file in files:
            # Check for valid image file extensions
            if file.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tiff")):
                file_path = os.path.join(root, file)
                try:
                    with Image.open(file_path) as img:
                        size = img.size  # Get image dimensions (width, height)
                        image_sizes[size] = image_sizes.get(size, 0) + 1
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")
    return image_sizes
```
