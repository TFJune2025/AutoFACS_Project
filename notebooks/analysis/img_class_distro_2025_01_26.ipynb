```python
from collections import Counter
import os
```

```python
def check_class_distribution(dataset_dir):
    """
    Counts the number of images per class in the dataset directory.
    Assumes the dataset is organized as:
    dataset_dir/
        class1/
            img1.jpg
            img2.jpg
        class2/
            img3.jpg
            img4.jpg
    """
    class_counts = Counter()
    for class_name in os.listdir(dataset_dir):
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path):
            num_images = len([
                f for f in os.listdir(class_path)
                if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp'))
            ])
            class_counts[class_name] += num_images
    
    print("Class Distribution:")
    for class_name, count in class_counts.items():
        print(f"{class_name}: {count} images")

    return class_counts
```

```python
# Example usage:
DATASET_DIR = "/home/natalyagrokh/img_datasets/combo_ferckja_dataset"
class_distribution = check_class_distribution(DATASET_DIR)
```
