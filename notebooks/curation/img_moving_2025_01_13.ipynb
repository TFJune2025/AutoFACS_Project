```python
import os
import shutil
```

```python
# Define the source and target directories
source_directory = "/Volumes/JavaAOT/Documents/AI/ml_expressions/img_datasets/celeba_dataset/extracted_images/img_align_celeba_png"
target_directory = "/Volumes/JavaAOT/Documents/AI/ml_expressions/img_datasets/celeba_dataset"
```

```python
# Ensure the target directory exists
os.makedirs(target_directory, exist_ok=True)
```

```python
# Define the list of image file extensions to look for
image_extensions = (".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff")
```

```python
# Walk through the source directory and find image files
for root, dirs, files in os.walk(source_directory):
    for file in files:
        # Check if the file is an image
        if file.lower().endswith(image_extensions):
            source_file = os.path.join(root, file)
            target_file = os.path.join(target_directory, file)
            
            # Move the file to the target directory
            try:
                shutil.move(source_file, target_file)
                print(f"Moved: {source_file} -> {target_file}")
            except Exception as e:
                print(f"Failed to move {source_file}: {e}")

print("All image files have been moved!")
```
