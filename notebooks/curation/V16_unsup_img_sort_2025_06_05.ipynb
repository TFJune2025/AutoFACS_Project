```python
import os
import shutil
```

```python
# --- USER CONFIG ---
v16_unsup_dir = "/Users/natalyagrokh/AI/ml_expressions/img_expressions/V16_unsup_20250604_142640/V16_unsup_20250604_142640_cluster_browser"
source_dir = "/Users/natalyagrokh/AI/ml_expressions/img_datasets/temp_autogluon_images"
destination_root = "/Users/natalyagrokh/AI/ml_expressions/img_datasets/unsup_training_clusters/V16_unsup"
```

```python
# --- MAIN SCRIPT ---
# Step 1: Create the destination skeleton
for cluster_folder in os.listdir(v16_unsup_dir):
    cluster_path = os.path.join(v16_unsup_dir, cluster_folder)
    if not os.path.isdir(cluster_path):
        continue
    dest_cluster_path = os.path.join(destination_root, cluster_folder)
    os.makedirs(dest_cluster_path, exist_ok=True)

    # ---- Copy images in the cluster root ----
    for img_file in os.listdir(cluster_path):
        full_img_path = os.path.join(cluster_path, img_file)
        if os.path.isdir(full_img_path) and img_file == "centroid_faces":
            # ---- Handle centroid_faces subfolder ----
            centroid_folder = os.path.join(full_img_path)
            dest_centroid_folder = os.path.join(dest_cluster_path, "centroid_faces")
            os.makedirs(dest_centroid_folder, exist_ok=True)
            for centroid_img in os.listdir(centroid_folder):
                if not centroid_img.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp')):
                    continue
                orig_path = os.path.join(source_dir, centroid_img)
                dest_path = os.path.join(dest_centroid_folder, centroid_img)
                if os.path.exists(orig_path):
                    shutil.copy2(orig_path, dest_path)
                else:
                    print(f"Centroid WARNING: {centroid_img} not found in source_dir!")
        elif img_file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp')):
            orig_path = os.path.join(source_dir, img_file)
            dest_path = os.path.join(dest_cluster_path, img_file)
            if os.path.exists(orig_path):
                shutil.copy2(orig_path, dest_path)
            else:
                print(f"WARNING: {img_file} not found in source_dir!")

print("Done! High-res clusters (including centroid_faces) created at:", destination_root)
```
