```python
# Recursively traverse all subdirectories and files in the source directory using os.walk
for root, dirs, files in os.walk(src_dir):
    # Loop over each file found in the current directory
    for file in files:
        src_file_path = os.path.join(root, file)  # Full path to the current file
        
        # Construct the new file path in the destination folder (flatten the structure)
        dst_file_path = os.path.join(dst_dir, file)
        
        # If a file with the same name already exists in the destination, append a counter to avoid overwriting
        base, ext = os.path.splitext(file)  # Separate the file name and its extension
        counter = 1  # Initialize counter for duplicate handling
        while os.path.exists(dst_file_path):
            # Create a new filename by appending the counter (e.g., 'file_1.ext')
            dst_file_path = os.path.join(dst_dir, f'{base}_{counter}{ext}')
            counter += 1
        
        # Copy the file from source to destination while preserving metadata
        shutil.copy2(src_file_path, dst_file_path)
        
        # Log the copy operation
        print(f'Copied: {src_file_path} -> {dst_file_path}')
```
