```python
import shutil
import os

def copy_folder(source_folder, destination_folder):
    """
    Copies a folder and its contents to a new location.

    Args:
        source_folder (str): Path to the source folder.
        destination_folder (str): Path to the destination folder.
    """
    try:
        if not os.path.exists(source_folder):
            print(f"Source folder does not exist: {source_folder}")
            return
        
        if os.path.exists(destination_folder):
            print(f"Destination folder already exists: {destination_folder}")
            return
        
        shutil.copytree(source_folder, destination_folder)
        print(f"Successfully copied {source_folder} to {destination_folder}")
    except Exception as e:
        print(f"Error copying folder: {e}")

# Example usage
source_folder = "/home/natalyagrokh/img_datasets/combo_ferckja_dataset"
destination_folder = "/home/natalyagrokh/img_datasets/combo_ferckja_dataset_2"

copy_folder(source_folder, destination_folder)
```
