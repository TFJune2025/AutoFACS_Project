```python
import os
import shutil
```

```python
# Define the source and target directories
source_directory = "/home/natalyagrokh/img_datasets/curated_images/home/natalyagrokh/img_datasets/ck_dataset_filtered"
target_directory = "/home/natalyagrokh/img_datasets/curated_images"
```

```python
# Ensure the target directory exists
os.makedirs(target_directory, exist_ok=True)
```

```python
# Walk through the source directory and move each folder
for folder_name in os.listdir(source_directory):
    folder_path = os.path.join(source_directory, folder_name)
    target_path = os.path.join(target_directory, folder_name)
    
    # Check if the item is a folder
    if os.path.isdir(folder_path):
        try:
            # Move the folder to the target directory
            shutil.move(folder_path, target_path)
            print(f"Moved folder: {folder_path} -> {target_path}")
        except Exception as e:
            print(f"Failed to move folder {folder_path}: {e}")

print("All folders have been moved!")
```
