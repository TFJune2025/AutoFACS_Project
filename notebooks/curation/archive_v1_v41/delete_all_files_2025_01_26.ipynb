```python
import os
```

```python
#code deletes all files in a folder
def delete_all_files(folder_path):
    """
    Delete all files in the specified folder.

    Args:
        folder_path (str): Path to the folder where files will be deleted.
    """
    if not os.path.exists(folder_path):
        print(f"Folder does not exist: {folder_path}")
        return

    files_deleted = 0

    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)

        if os.path.isfile(file_path):  # Ensure it's a file
            try:
                os.remove(file_path)  # Delete file
                files_deleted += 1
            except Exception as e:
                print(f"Error deleting {file_path}: {e}")

    print(f"Deleted {files_deleted} files from {folder_path}.")
```

```python
# Example usage
folder_path = "/home/natalyagrokh/img_datasets/combined_datasets_small_skipped"
delete_all_files(folder_path)
```
