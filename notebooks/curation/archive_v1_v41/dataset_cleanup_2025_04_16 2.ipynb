```python
#CODE VERSION #0 ->

from PIL import Image
import imagehash
from pathlib import Path
import os

IMG_DIR = "/Volumes/JavaAOT/Documents/AI/ml_expressions/img_datasets/ferckjalf_dataset/happiness (sort with perfect future model)"
hashes = {}
duplicates = []

HAMMING_THRESHOLD = 5

for img_path in Path(IMG_DIR).rglob("*.[jpJP][pnPN]*[gG]"):
    try:
        img = Image.open(img_path).convert("RGB")
        new_hash = imagehash.phash(img)

        for h, stored_path in hashes.items():
            if h - new_hash < HAMMING_THRESHOLD:
                duplicates.append((img_path, stored_path))  # (duplicate, original)
                break
        else:
            hashes[new_hash] = img_path

    except Exception as e:
        print(f"Failed to process {img_path}: {e}")

# Delete duplicates and print their source
for dup_path, original_path in duplicates:
    try:
        os.remove(dup_path)
        print(f"üóëÔ∏è Deleted: {dup_path}")
        print(f"   ‚Ü™Ô∏è Similar to: {original_path}\n")
    except Exception as e:
        print(f"‚ùå Failed to delete {dup_path}: {e}")

print(f"\n‚úÖ Deduplication complete. {len(duplicates)} duplicates removed.")
```

```python
#CODE VERSION #1-> check if it's the same as above
from PIL import Image
import imagehash
from pathlib import Path
import os

IMG_DIR = "/Volumes/JavaAOT/Documents/AI/ml_expressions/img_datasets/ferckjalf_dataset/neutral copy"
hashes = {}
duplicates = []

# Threshold for near-duplicate detection (lower = stricter)
HAMMING_THRESHOLD = 5

# .rglob method from pathlib module 
    #recursively searches all subdirectories for matching files
for img_path in Path(IMG_DIR).rglob("*.[jpJP][pnPN]*[gG]"):
    try:
        img = Image.open(img_path).convert("RGB")
        new_hash = imagehash.phash(img)

        for h, stored_path in hashes.items():
            if h - new_hash < HAMMING_THRESHOLD:
                duplicates.append((img_path, stored_path))  # (duplicate, original)
                break
        else:
            hashes[new_hash] = img_path

    except Exception as e:
        print(f"Failed to process {img_path}: {e}")

# Delete duplicates and print their source
for dup_path, original_path in duplicates:
    try:
        os.remove(dup_path)
        print(f"üóëÔ∏è Deleted: {dup_path}")
        print(f"   ‚Ü™Ô∏è Similar to: {original_path}\n")
    except Exception as e:
        print(f"‚ùå Failed to delete {dup_path}: {e}")

print(f"\n‚úÖ Deduplication complete. {len(duplicates)} duplicates removed.")
```
