```python
import os
import shutil
```

```python
# Define the base source directory where Img_1 to Img_15 are located
base_directory = "/home/natalyagrokh/img_datasets"
```

```python
# Define the target directory to consolidate all .7z files
target_directory = "/home/natalyagrokh/img_datasets/celeba_dataset"
```

```python
# Ensure the target directory exists
os.makedirs(target_directory, exist_ok=True)
```

```python
# Keep track of files moved
files_moved = 0
```

```python
# Loop through Img_1 to Img_15 and handle nested structures
for i in range(1, 16):  # Iterate through Img_1 to Img_15
    source_root = os.path.join(base_directory, f"Img_{i}")
    print(f"Checking source root: {source_root}")  # Debugging line
    
    # Check if the source directory exists
    if not os.path.exists(source_root):
        print(f"Source directory does not exist: {source_root}")
        continue

    # Walk through all directories and subdirectories
    for root, dirs, files in os.walk(source_root):
        for file in files:
            # Debugging: Print the file name being evaluated
            print(f"Found file: {file}")
            
            # Check if the file matches the pattern for split archives
            if file.lower().startswith("img_align_celeba_png.7z."):
                # Construct full file paths
                source_file = os.path.join(root, file)
                target_file = os.path.join(target_directory, file)

                print(f"Attempting to move: {source_file} -> {target_file}")  # Debugging
                
                try:
                    # Move the file to the target directory
                    shutil.move(source_file, target_file)
                    print(f"Moved: {source_file} -> {target_file}")
                    files_moved += 1
                except Exception as e:
                    print(f"Failed to move {source_file}: {e}")
```

```python
print(f"Total files moved: {files_moved}")
if files_moved == 0:
    print("No .7z files were found or moved. Please check the folder structure and file extensions.")
```
