```python
import zipfile
import os

# Path to the zip file
zip_file_path = "/home/natalyagrokh/img_datasets/archive (3).zip"
# Directory where the contents will be extracted
output_dir = "/home/natalyagrokh/img_datasets/faces"

try:
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)
    
    # Open the zip file and extract all contents
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        print(f"Extracting contents of {zip_file_path} to {output_dir}...")
        zip_ref.extractall(output_dir)
    
    print(f"Extraction completed. Files are in {output_dir}")
except FileNotFoundError:
    print(f"Error: The file {zip_file_path} does not exist.")
except zipfile.BadZipFile:
    print(f"Error: {zip_file_path} is not a valid zip file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
