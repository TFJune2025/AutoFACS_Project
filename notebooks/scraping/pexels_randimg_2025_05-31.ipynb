```python
import requests
import os
import time
from dotenv import load_dotenv
```

```python
UNKNOWN_DIR = "/Users/natalyagrokh/AI/ml_expressions/img_datasets/ferckjalf_dataset/unknown"
os.makedirs(UNKNOWN_DIR, exist_ok=True)

# Load variables from .env
load_dotenv()

PEXELS_API_KEY = os.getenv("PEXELS_API_KEY")

if not PEXELS_API_KEY:
    raise ValueError("PEXELS_API_KEY not found in .env file")
```

```python
def download_pexels_images(n=100):
    headers = {"Authorization": PEXELS_API_KEY}
    page = 1
    downloaded = 0
    while downloaded < n:
        url = f"https://api.pexels.com/v1/curated?per_page=15&page={page}"
        resp = requests.get(url, headers=headers)
        data = resp.json()
        photos = data.get("photos", [])
        if not photos:
            break
        for photo in photos:
            img_url = photo["src"]["medium"]
            try:
                img_data = requests.get(img_url).content
                fname = os.path.join(UNKNOWN_DIR, f"pexels_{downloaded}.jpg")
                with open(fname, "wb") as f:
                    f.write(img_data)
                downloaded += 1
                if downloaded >= n:
                    break
            except Exception as e:
                print(f"Failed to download: {img_url} | {e}")
        page += 1
        time.sleep(0.7)  # polite delay
    print(f"✅ Downloaded {downloaded} Pexels images to {UNKNOWN_DIR}")
```

```python
# Download 100 random Pexels images
download_pexels_images(500)
```
